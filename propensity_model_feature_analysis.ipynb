```python
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
df = pd.read_csv('propensity_dataset.csv')
```

```python
resp_df = df[df['responded'] == 1].reset_index(drop = True)
```

```python
funded_df = df[df['funded'] == 1].reset_index(drop = True)
```

```python
df.columns
```

```python
feature_list = ['Tenure', 'Average_Turnover', 'Average_Trans_Count',
       'Times_Marketed_To', 'Message', 'Xer', 'mcc_rating', 'Pam_Score']
```

## Plotting Response/Fund Rates

```python
def response_plot(df, feature, n, rate_col, pop):
    if rate_col == 'responded':
        rate = 'Response'
        max_y = 0.01
    elif rate_col == 'funded':
        rate = 'Fund'
        max_y = 1
    df[f'{feature}_{n}tile'] = pd.qcut(df[feature], n, labels = False)
    df[[f'{feature}_{n}tile', rate_col]].groupby(f'{feature}_{n}tile')[[rate_col]].mean().plot(ylim = (0, max_y), xticks = range(1, n, 1), title = f'{rate} rate of {pop} by {feature}')
```

```python
for a in range(1, 6, 1):
    print(a)
```

```python
response_plot(df, feature_list[5], 5, 'responded', 'all mailings')
```

# What % of each risk group respond & fund?

### Response rate by PAM score

```python
response_plot(df,'Pam_Score', 5, 'responded', 'all mailings')
```

### Fund Rate of responses by PAM score

```python
response_plot(resp_df, 'Pam_Score', 5, 'funded', 'responses')
```

# Response and Fund Rate by industry, by month

### Response Rate

```python
df.groupby(['land_month_no', 'Industry'])['responded'].mean().reset_index().pivot(index = 'land_month_no', columns = 'Industry', values = 'responded').plot(title = 'Response rate by Month Number, by Industry', figsize = (15, 5))
```

### Fund Rate

```python
resp_df.groupby(['land_month_no', 'Industry'])['funded'].mean().reset_index().pivot(index = 'land_month_no', columns = 'Industry', values = 'funded').plot(ylim = (0, 1), title = 'Fund rate by Month Number, by Industry', figsize = (15, 5))
```
