```python
from fast_to_SQL.fast_to_SQL import to_sql_fast
import pandas as pd
from urllib.parse import quote_plus
import sqlalchemy

params = quote_plus(
    """DRIVER={ODBC Driver 13 for SQL Server};
    SERVER=server-dw01.database.windows.net;
    DATABASE=DataWarehouseStaging;
    UID=dw_userlogin;
    PWD=kljDueK!7"""
)

ENGINE = sqlalchemy.create_engine("mssql+pyodbc:///?odbc_connect=%s" % params)
```

```python
Xer_df = pd.read_sql(''' SELECT * FROM pam.dataset_Xer ''', ENGINE)
```

```python
df = pd.melt(Xer_df, id_vars = ['com_id_h'], var_name = 'month', value_name = 'Xer')
```

```python
mktg_df = pd.read_sql(''' SELECT [Hashed_ID], [Land_Date], [Opportunity ID], [MCC_Description]
                            FROM [marketing].[WP_Attri_Mailers]
                            WHERE Marketed_To = 1''', ENGINE)
```

```python
mktg_df.head()
```

```python
df.head()
```

```python
month_no_dict = {'jan' : 1,
'feb' : 2,
'mar' : 3,
'apr' : 4,
'may' : 5,
'jun' : 6,
'jul' : 7,
'aug' : 8,
'sep' : 9,
'oct' : 10,
'nov' : 11,
'dec' : 12
}
```

```python
df['month_no'] = df['month'].map(month_no_dict)
df['month_no_+1'] = df['month'].map(month_no_dict).add(1).replace(13, 1)
df['month_no_-1'] = df['month'].map(month_no_dict).add(-1).replace(0, 12)
```

```python
df.head()
```

```python
mktg_df['Land_Date'] = pd.to_datetime(mktg_df['Land_Date'])
```

```python
mktg_df['land_month'] = pd.DatetimeIndex(mktg_df['Land_Date']).month
```

```python
import calendar
mktg_df['land_month'] = mktg_df['land_month'].apply(lambda x: calendar.month_abbr[x])
```

```python
mktg_df['land_month'] = mktg_df['land_month'].str.lower()
```

```python
mktg_df['land_month_no'] = mktg_df['land_month'].map(month_no_dict)
mktg_df['land_month_no_+1'] = mktg_df['land_month'].map(month_no_dict).add(1).replace(13, 1)
mktg_df['land_month_no_-1'] = mktg_df['land_month'].map(month_no_dict).add(-1).replace(0, 12)
```

```python
mktg_df.head()
```

```python
mktg_df['responded'] = mktg_df['Opportunity ID'].apply(lambda x: 0 if pd.isnull(x) else 1)
```

```python
%matplotlib inline
```

```python
final_df = mktg_df.merge(right = df, how = 'left', left_on = ['Hashed_ID','land_month_no'], right_on = ['com_id_h', 'month_no'])

final_df = final_df[['com_id_h', 'MCC_Description', 'month', 'Xer', 'responded']]

final_df['overall_decile'] = pd.qcut(final_df['Xer'], 10, labels = False)

normal_df = final_df.sort_values('Xer', ascending = False).iloc[20:]
normal_df.groupby('overall_decile')['responded', 'Xer'].mean().plot(secondary_y = ['Xer'], ylim = (0, 0.005))
```

```python
final_df
```

```python
mcc_df = pd.read_sql('''SELECT * FROM pam.MCC_cluster_dict''', ENGINE)
```

```python
mcc_df.head()
```

```python
industry_df = final_df.merge(mcc_df, how = 'left', left_on = 'MCC_Description', right_on = 'MCC_DESC')
```

```python
industry_df = industry_df[['com_id_h', 'Xer', 'responded', 'overall_decile', 'Dominant_Topic', 'MCC_Description']]
```

```python
industry_df.to_csv('industry_df.csv', index = False)
```
